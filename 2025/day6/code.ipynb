```python
test, real = [], []
with open("./test.txt", "r+") as f:
    test = f.readlines()

with open("./real.txt", "r+") as f:
    real = f.readlines()
```

```python
import math

def part1(lines: list[str]):
    s = 0
    n_rows = len(lines) - 1

    grid = [list() for _ in range(n_rows)]
    ops = []
    word = ""
    for idx, line in enumerate(lines[:-1]):
        for c in line:
            if (c == " " or c == "\n") and len(word) == 0:
                continue
            elif c == " " or c == "\n":
                grid[idx].append(int(word))
                word = ""
            else:
                word += c
    
    for c in lines[-1]:
        if c == " ":
            continue
        ops.append(c)
    
    n_cols = len(grid[0])
    for col in range(n_cols):
        col_nums = []
        for idx in range(n_rows):
            col_nums.append(grid[idx][col])
        
        if ops[col] == "+":
            s += sum(col_nums)
        else:
            s += math.prod(col_nums)

    return s

print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python
def part2(lines: list[str]):
    s = 0
    buf = []
    for idx in reversed(range(len(lines[0]) - 1)):
        word = ""
        for row in range(len(lines)):
            c = lines[row][idx]
            if c not in [' ', '+', '*']:
                word += c
            elif c != ' ':
                buf.append(int(word))
                if c == "+":
                    s += sum(buf)
                else:
                    s += math.prod(buf)
                buf = []
                word = ""
        if len(word) > 0:
            buf.append(int(word))

    return s

print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
