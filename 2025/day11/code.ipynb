```python
from itertools import product
from collections import defaultdict
import math
from tqdm import tqdm
import re
import numpy as np

test, real = [], []
p2_test = []
with open("./test.txt", "r+") as f:
    test = [l.rstrip() for l in f.readlines()]

with open("./real.txt", "r+") as f:
    real = [l.rstrip() for l in f.readlines()]

with open("./part2_test.txt", "r+") as f:
    p2_test = [l.rstrip() for l in f.readlines()]
```

```python
from collections import deque
def part1(lines: list[str]):
    adj = {}
    for line in lines:
        name = line[:3]
        adjs = line[5:].split(' ')
        adj[name] = adjs
    
    flow = {k: 0 for k in adj.keys()}
    flow['out'] = 0
    flow['you'] += 1

    d = deque(['you'])
    while len(d) > 0:
        cur = d.popleft()
        if cur == 'out': continue

        for n in adj[cur]:
            flow[n] += 1
            d.append(n)

    return flow['out']
print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python
from functools import cache

def part2(lines: list[str]):
    adj = {}
    for line in lines:
        name = line[:3]
        adjs = line[5:].split(' ')
        adj[name] = adjs
    
    @cache
    def paths(cur: str = "svr", fft: bool = False, dac: bool = False):
        if cur == "out":
            return 1 if (fft and dac) else 0
        
        has_fft = fft or cur == "fft"
        has_dac = dac or cur == "dac"
        return sum([
             paths(n, has_fft, has_dac) for n in adj[cur]
        ])
        
    return paths()
        
print("TEST: ", part2(p2_test))
print("REAL: ", part2(real))
```
