```python
from itertools import product
test, real = [], []
with open("./test.txt", "r+") as f:
    test = [l.rstrip() for l in f.readlines()]

with open("./real.txt", "r+") as f:
    real = [l.rstrip() for l in f.readlines()]
```

```python

def d(a: list[int], b: list[int]) -> int:
    return sum([abs(i - j)**2 for i,j in zip(a,b)])

from collections import defaultdict
import math

def find_root(x: int, d: dict) -> int:
    cur = x
    while d[cur][0] is not None:
        cur = d[cur][0]
    return cur
    

def part1(lines: list[str], m: int):
    nums = [
        list(map(int, ns.split(','))) for ns in lines
    ]

    pair_ordering = []
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            pair_ordering.append((d(nums[i], nums[j]), i, j))
    
    pair_ordering.sort()
    pair_ordering = pair_ordering[:m]

    # union-find
    u_set = defaultdict(lambda: (None, 1))

    for _, i, j in pair_ordering:
        ir, jr = find_root(i, u_set), find_root(j, u_set)

        if ir == jr:
            continue
        
        if u_set[ir][1] < u_set[jr][1]:
            jr, ir = ir, jr
        
        u_set[ir] = (None, u_set[ir][1] + u_set[jr][1])
        u_set[jr] = (ir, u_set[jr][1])
    
    sizes = []
    for k, v in u_set.items():
        if v[0] is None:
            sizes.append(v[1])
    sizes.sort(reverse=True)
    return math.prod(sizes[:3])

print("TEST: ", part1(test, 10))
print("REAL: ", part1(real, 1000))
```

```python
def part2(lines: list[str]):
    nums = [
        list(map(int, ns.split(','))) for ns in lines
    ]

    pair_ordering = []
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            pair_ordering.append((d(nums[i], nums[j]), i, j))
    
    pair_ordering.sort()

    # union-find
    u_set = defaultdict(lambda: (None, 1))

    for _, i, j in pair_ordering:
        ir, jr = find_root(i, u_set), find_root(j, u_set)

        if ir == jr:
            continue
        
        if u_set[ir][1] < u_set[jr][1]:
            jr, ir = ir, jr
        
        u_set[ir] = (None, u_set[ir][1] + u_set[jr][1])
        u_set[jr] = (ir, u_set[jr][1])

        if u_set[ir][1] == len(lines):
            return nums[i][0] * nums[j][0]
    
print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
