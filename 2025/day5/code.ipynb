```python
test, real = [], []
with open("./test.txt", "r+") as f:
    test = f.readlines()

with open("./real.txt", "r+") as f:
    real = f.readlines()
```

```python
def part1(lines: list[str]):
    ranges = []
    l_idx = 0
    for idx, line in enumerate(lines):
        line = line.strip()
        
        if len(line) == 0:
            l_idx = idx
            break

        spl = line.split('-')
        l, r = int(spl[0]), int(spl[1])
        ranges.append((l, r))
    
    queries = []
    for line in lines[l_idx + 1:]:
        queries.append(int(line))
    
    s = 0
    for q in queries:
        for (l, r) in ranges:
            if l <= q and q <= r:
                s += 1
                break
    return s

print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python
def part2(lines: list[str]):
    ranges = []
    for idx, line in enumerate(lines):
        line = line.strip()
        
        if len(line) == 0:
            break

        spl = line.split('-')
        l, r = int(spl[0]), int(spl[1])
        ranges.append((l, r))
       
    events = []
    for l, r in ranges:
        events.extend([(l, 1), (r, -1)])
    events.sort(key=lambda x: (x[0], -x[1]))

    cur_idx = events[0][0]
    c = 1
    s = 1
    for idx, d in events[1:]:
        if c == 0 and d == 1:
            s += 1
        else:
            s += (idx - cur_idx)
        
        c += d
        
        cur_idx = idx
            
    return s

print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
