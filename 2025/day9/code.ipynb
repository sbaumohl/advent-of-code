```python
from itertools import product
from collections import defaultdict
import math
from tqdm import tqdm

test, real = [], []
with open("./test.txt", "r+") as f:
    test = [l.rstrip() for l in f.readlines()]

with open("./real.txt", "r+") as f:
    real = [l.rstrip() for l in f.readlines()]
```

```python
def part1(lines: list[str]):
    nums = [list(map(int, l.split(','))) for l in lines]
    m = -1
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            d = (abs(nums[i][0] - nums[j][0]) + 1) * (abs(nums[i][1] - nums[j][1]) + 1)
            m = max(m ,d)

    return m

print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python

# assumption, no lines are diagonal.
# !pip install shapely
from functools import partial
from shapely.geometry import Point, box
from shapely.geometry.polygon import Polygon

def part2(lines: list[str]):
    nums = [list(map(int, l.split(','))) for l in lines]
    polygon = Polygon([(x[0], x[1]) for x in nums]) 
    s = 0
    for i in tqdm(range(len(nums))):
        a = nums[i]
        for b in nums[i + 1:]:
            rect = box(min(a[0], b[0]), min(a[1], b[1]), max(a[0], b[0]), max(a[1], b[1]))

            if polygon.covers(rect):
                s = max(s, (abs(a[0] - b[0]) + 1) * (abs(a[1] - b[1]) + 1))
                
    return s
        
print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
