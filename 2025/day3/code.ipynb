```python
test, real = [], []
with open("./day3/test.txt", "r+") as f:
    test = f.readlines()

with open("./day3/real.txt", "r+") as f:
    real = f.readlines()
```

```python
import numpy as np
def part1(lines: list[str]):
    c = 0
    for line in lines:
        jolts = list(map(int, line.strip()))
        l_idx = np.argmax(jolts[:-1])
        r = max(jolts[l_idx + 1:])
        l = jolts[l_idx]
        c += int(f"{l}{r}")
    return c

print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python
def part2(lines: list[str]):
    c = 0
    for line in lines:
        jolts = list(map(int, line.strip()))
        s = []
        last_idx = -1
        for digit in range(12):
            digits_left = 11 - digit
            end_idx = -digits_left if digits_left != 0 else len(jolts)
            last_idx = np.argmax(jolts[last_idx + 1:end_idx]) + (last_idx + 1)
            s.append(str(jolts[last_idx]))
        c += int(''.join(s))
    return c

print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
