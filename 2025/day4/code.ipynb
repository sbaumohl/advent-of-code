```python
test, real = [], []
with open("./test.txt", "r+") as f:
    test = f.readlines()

with open("./real.txt", "r+") as f:
    real = f.readlines()
```

```python
from itertools import product
def part1(lines: list[str]):
    lines = [l.strip() for l in lines]
    s = 0
    for i, j in product(range(len(lines)), range(len(lines[0]))):
        c = 0
        for di, dj in product([1, -1, 0], [1, -1, 0]):
            ni, nj = i + di, j + dj
            if ni < 0 or ni >= len(lines):
                continue
            elif nj < 0 or nj >= len(lines[0]):
                continue
            if di == 0 and dj == 0: 
                continue
            if lines[ni][nj] == '@':
                c += 1
        
        if c < 4 and lines[i][j] == '@':
            s += 1 
    return s
    

print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python
def part2(lines: list[str]):
    lines = [[_ for _ in l.strip()] for l in lines]
    s = 0
    last_round = 1
    while last_round > 0:
        last_round = 0
        for i, j in product(range(len(lines)), range(len(lines[0]))):
            c = 0
            for di, dj in product([1, -1, 0], [1, -1, 0]):
                ni, nj = i + di, j + dj
                if ni < 0 or ni >= len(lines):
                    continue
                elif nj < 0 or nj >= len(lines[0]):
                    continue
                if di == 0 and dj == 0: 
                    continue
                if lines[ni][nj] == '@':
                    c += 1
            
            if c < 4 and lines[i][j] == '@':
                lines[i][j] = 'x'
                s += 1 
                last_round += 1
    return s

print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
