```python
test, real = [], []
with open("./test.txt", "r+") as f:
    test = [l.rstrip() for l in f.readlines()]

with open("./real.txt", "r+") as f:
    real = [l.rstrip() for l in f.readlines()]
```

```python

def part1(lines: list[str]):
    s = 0
    beams = [False] * len(lines[0])
    for idx, c in enumerate(lines[0]):
        if c == 'S':
            beams[idx] = True
    
    for idx, row in enumerate(lines, 1):
        buf = list(beams)

        for cidx, c in enumerate(row):
            if c == '^' and beams[cidx]:
                buf[cidx] = False
                if cidx != 0:
                    buf[cidx - 1] = True
                if cidx != len(row) - 1:
                    buf[cidx + 1] = True
                s += 1
        
        beams = buf
    return s

print("TEST: ", part1(test))
print("REAL: ", part1(real))
```

```python
def part2(lines: list[str]):
    beams = [0] * len(lines[0])
    for idx, c in enumerate(lines[0]):
        if c == 'S':
            beams[idx] = 1
            break
    
    for idx, row in enumerate(lines, 1):
        buf = list(beams)

        for cidx, c in enumerate(row):
            if c == '^' and beams[cidx]:
                t = buf[cidx]
                buf[cidx] = 0
                if cidx != 0:
                    buf[cidx - 1] += t
                if cidx != len(row) - 1:
                    buf[cidx + 1] += t
        beams = buf
    return sum(beams)

print("TEST: ", part2(test))
print("REAL: ", part2(real))
```
